```python
import random

word_list = [
    "quirky", "bizarre", "melancholy", "whimsical", "peculiar", "luminous", "enigmatic",
    "serene", "vivid", "eccentric", "mysterious", "dazzling", "flourish", "twinkle", "gleam",
    "baffling", "jovial", "radiant", "elusive", "vibrant", "nostalgic", "effervescent",
    "curious", "tranquil", "elixir", "splendid", "mosaic", "witty", "mystical", "twilight"
]

# 隨機選擇單詞
secret_word = random.choice(word_list)

# AI 猜測函數
def ai_guess():
    # 模擬AI隨機選字母猜測
    guess = random.choice('abcdefghijklmnopqrstuvwxyz')
    return guess

# 兩個AI互動並且進行4回合，並比賽誰猜得更多
def game():
    round_count = 0  # 計算回合次數
    ai1_score = 0  # AI 1 得分
    ai2_score = 0  # AI 2 得分
    game_data = []  # 用來保存遊戲過程
    game_data.append(["遊戲開始", f"出題者選定了一個 {len(secret_word)} 個字母的單詞！"])

    # 遊戲進行4回合
    while round_count < 4:
        round_count += 1
        game_data.append([f"回合 {round_count} 開始", f"出題者選定的單詞長度是 {len(secret_word)}。"])

        guessed_word = ["_"] * len(secret_word)  # 每回合的猜測初始化
        attempts = 5  # 每回合最多可錯 5 次
        wrong_guesses = []

        # AI 1 猜測
        ai1_guesses = 0
        while "_" in guessed_word and attempts > 0:
            guess = ai_guess()  # AI 1 猜測
            game_data.append([f"AI 1 猜測", guess])

            if guess in secret_word and guess not in guessed_word:
                # 替換正確的字母
                for i, letter in enumerate(secret_word):
                    if letter == guess:
                        guessed_word[i] = guess
                ai1_guesses += 1
                game_data.append([f"AI 1 正確！目前單詞", " ".join(guessed_word)])
            else:
                attempts -= 1
                wrong_guesses.append(guess)
                game_data.append([f"錯誤！剩餘機會", attempts])

        # AI 2 猜測
        ai2_guesses = 0
        attempts = 5  # 重置AI 2的嘗試次數
        wrong_guesses = []

        while "_" in guessed_word and attempts > 0:
            guess = ai_guess()  # AI 2 猜測
            game_data.append([f"AI 2 猜測", guess])

            if guess in secret_word and guess not in guessed_word:
                # 替換正確的字母
                for i, letter in enumerate(secret_word):
                    if letter == guess:
                        guessed_word[i] = guess
                ai2_guesses += 1
                game_data.append([f"AI 2 正確！目前單詞", " ".join(guessed_word)])
            else:
                attempts -= 1
                wrong_guesses.append(guess)
                game_data.append([f"錯誤！剩餘機會", attempts])

        # 結算每回合得分
        if ai1_guesses > ai2_guesses:
            ai1_score += 1
            game_data.append([f"回合 {round_count} 結束", f"AI 1 獲勝！本回合 AI 1 猜對了 {ai1_guesses} 個字母，AI 2 猜對了 {ai2_guesses} 個字母。"])
        elif ai2_guesses > ai1_guesses:
            ai2_score += 1
            game_data.append([f"回合 {round_count} 結束", f"AI 2 獲勝！本回合 AI 2 猜對了 {ai2_guesses} 個字母，AI 1 猜對了 {ai1_guesses} 個字母。"])
        else:
            game_data.append([f"回合 {round_count} 結束", f"本回合平手！AI 1 和 AI 2 猜對了相同數量的字母。"])

    # 結算總結果
    if ai1_score > ai2_score:
        game_data.append([f"遊戲結束", f"AI 1 總得分 {ai1_score}，AI 2 總得分 {ai2_score}，AI 1 獲勝！"])
    elif ai2_score > ai1_score:
        game_data.append([f"遊戲結束", f"AI 1 總得分 {ai1_score}，AI 2 總得分 {ai2_score}，AI 2 獲勝！"])
    else:
        game_data.append([f"遊戲結束", f"平手！AI 1 和 AI 2 得分相同。"])

    return game_data

# 開始遊戲並印出結果
game_result = game()

# 顯示遊戲過程
for entry in game_result:
    print(entry[0], ":", entry[1])
```

```python
df_game_data = pd.DataFrame(game_result, columns=["事件", "結果"])

# 顯示 DataFrame
print(df_game_data)
```
